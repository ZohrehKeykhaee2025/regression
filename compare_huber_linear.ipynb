# library

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import HuberRegressor, LinearRegression
from sklearn.metrics import mean_squared_error
```

# data

```python
np.random.seed(0)
X = np.random.rand(100, 1) * 10
y = 2.5 * X.squeeze() + np.random.randn(100) * 2
y[::10] += 15  # آوتلایر

X_test = np.linspace(0, 10, 100).reshape(-1, 1)
```

# models

```python
huber = HuberRegressor()
linear = LinearRegression()

huber.fit(X, y)
linear.fit(X, y)

y_huber = huber.predict(X_test)
y_linear = linear.predict(X_test)
```

# plot

```python
plt.scatter(X, y, alpha=0.4, label='Data')
plt.plot(X_test, y_huber, color='red', label='Huber Regression')
plt.plot(X_test, y_linear, color='blue', linestyle='--', label='Linear Regression')
plt.title('Huber vs Linear Regression')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

# MSE

```python
print("MSE - Huber:", mean_squared_error(y, huber.predict(X)))
print("MSE - Linear:", mean_squared_error(y, linear.predict(X)))
```
