# library

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import HuberRegressor
```

# data

```python
np.random.seed(0)
X = np.random.rand(100, 1) * 10
y = 2.5 * X.squeeze() + np.random.randn(100) * 2
y[::10] += 15  # اضافه کردن آوتلایر
```

# huber ragression

```python
model = HuberRegressor()
model.fit(X, y)
```

# production

```python
X_test = np.linspace(0, 10, 100).reshape(-1, 1)
y_pred = model.predict(X_test)
```

# plot

```python
plt.scatter(X, y, color='gray', label='Data with Outliers')
plt.plot(X_test, y_pred, color='red', label='Huber Regression')
plt.legend()
plt.title('Huber Regression Example')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```
